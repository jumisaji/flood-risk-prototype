This notebook trains a Logistic Regression classifier using the shared `common.py` module so that every model in the project uses the same dataset, feature engineering, chronological split and evaluation metrics.


```
import warnings
warnings.filterwarnings("ignore")

import joblib
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import ConfusionMatrixDisplay, RocCurveDisplay
import common
```

## 1. Load data and build features

```
df = common.load_data()

X, y = common.build_features(df)

X_train, X_test, y_train, y_test = common.chronological_split(X, y)

print(f"Training samples: {len(X_train)}")
print(f"Testing samples: {len(X_test)}")
```

## 2. Train Logistic Regression

```
model = LogisticRegression(
    class_weight="balanced",
    max_iter=1000,
    random_state=42
)

model.fit(X_train, y_train)
```

## 3. Predictions

```
preds = model.predict(X_test)
probs = model.predict_proba(X_test)[:,1]
```

## 4. Evaluate model

```
baseline = common.persistence_baseline(df, y_test)

results = [
    baseline,
    common.evaluate(
        "Logistic Regression",
        y_test,
        preds,
        probs
    )
]

comparison = common.comparison_table(results)
comparison
```

## 5. Confusion Matrix

```
ConfusionMatrixDisplay.from_predictions(y_test, preds)
plt.show()
```

## 6. ROC Curve

```
RocCurveDisplay.from_predictions(y_test, probs)
plt.show()
```

## 7. Save trained model

```
joblib.dump(model, "../models/logistic_regression.joblib")
print("Model saved to ../models/logistic_regression.joblib")
```

## Conclusions

- The model was trained using the shared project workflow.
- Evaluation metrics were generated using `common.evaluate()`.
- Performance can be compared directly against the persistence baseline.
